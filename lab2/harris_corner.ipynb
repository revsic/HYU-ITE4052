```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
from harris_corner import harris_matrix, harris_resp
```

```python
checkerboard = plt.imread('./checkerboard.png')
checkerboard.shape
```

```python
plt.imshow(checkerboard, cmap='gray')
```

```python
candidates = [3, 7, 11]
```

```python
base = 5
plt.figure(figsize=(3 * base, len(candidates) * base))

for i, size in enumerate(candidates):
    matrix = harris_matrix(checkerboard, size)
    eigval, _, resp = harris_resp(matrix)
    
    plt.subplot(len(candidates), 3, i * 3 + 1)
    plt.title('eigval min')
    plt.imshow(eigval.min(axis=-1))
    
    plt.subplot(len(candidates), 3, i * 3 + 2)
    plt.title('eigval max')
    plt.imshow(eigval.max(axis=-1))
    
    plt.subplot(len(candidates), 3, i * 3 + 3)
    plt.title('harris response')
    plt.imshow(resp)
```
